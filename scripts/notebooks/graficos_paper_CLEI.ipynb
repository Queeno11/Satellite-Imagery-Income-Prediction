```python
import pylab as pl
import numpy as np
from pylab import *

a = np.array([[0,10]])
pl.figure(figsize=(9, 0.5))
cmap = cm.get_cmap('Spectral', 10)    # 11 discrete colors
img = pl.imshow(a, cmap=cmap)
pl.gca().set_visible(False)
cax = pl.axes([0.1, 0.2, 0.8, 0.6])

cbar = pl.colorbar(orientation="horizontal", norm=mpl.colors.Normalize(0, 10), cax=cax)
cbar.set_label('Deciles del Ingreso per cápita del Censo 2010', fontsize=12)

# pl.savefig("colorbar.pdf")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

model = "effnet_v2S_lr0.0001_size128_y2013-2018-2022"
path = fr"D:\Maestría\Tesis\3-Resultados_post_estimacion\outputs\{model}"
df = pd.read_excel(rf"{path}\municipios_{model}.xlsx", header=1)

fig, ax = plt.subplots(2,2, figsize=[8,8])

row = 0
col = 0
indicators = ["Media", "Mediana", "Gini", "FGT(0) Nac 2010ppp"]
for ind in indicators:
    x = df[f"{ind}"]
    y = df[f"{ind}.1"]
    sns.regplot(x=x, y=y, ci=None, ax=ax[col][row])
    ax[col][row].plot([min(x), max(x)], [min(x), max(x)], linestyle='--', color='red')
    ax[col][row].set_title(f'{ind}')
    ax[col][row].set_xlabel("")
    ax[col][row].set_ylabel("")

    row += 1
    if row>=2:
        row=0
        col+=1
ax[1][0].set_xlabel('Datos censales')
ax[1][1].set_xlabel('Datos censales')
ax[0][0].set_ylabel('Predicciones del modelo')
ax[1][0].set_ylabel('Predicciones del modelo')

plt.tight_layout()        
fig.savefig(rf"{path}\municipios_paper.png")
```
